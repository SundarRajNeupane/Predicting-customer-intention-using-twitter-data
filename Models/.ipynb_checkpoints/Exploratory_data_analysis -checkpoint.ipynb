```python
# looking at the mean values for likes, retweets and replies to tweets sent on cost of living

print('-' * 80)
print(f'Who receives more likes? verified or non verified accounts?',df.groupby('verified').tweet_like_count.mean())
print('-' * 80)
print(f'Who receives more retweets? verified or non verified accounts?',df.groupby('verified').tweet_retweet_count.mean())
print('-' * 80)
print(f'Who receives more replies? verified or non verified accounts?',df.groupby('verified').tweet_reply_count.mean())
print('-' * 80)
print(f'Who has more followers? verified or non verified accounts?',df.groupby('verified').followers_count.mean())
print('-' * 80)
print(f'Who follows more users, verified or non verified accounts?',df.groupby('verified').following_count.mean())
print('-' * 80)
```

```python
stopwords_set = set(STOPWORDS)

def wordcloud(df, feature):
    wordcloud = WordCloud(background_color = 'black',
                         stopwords = stopwords_set,
                         width = 1000,
                         height = 800,
                         colormap = 'Set3',
                         margin = 0,
                         max_words = 300,
                         min_word_length = 4,
                         max_font_size = 130, min_font_size = 15,).generate(" ".join(df[feature]))
    plt.figure(figsize = (20,15))
    plt.imshow(wordcloud)
    plt.axis('off')
 
```

```python
stopword_tweet_text=wordcloud(df, 'tweet_text')
```

```python
stopwords_set = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'black',
                     stopwords = stopwords_set,
                     width = 1000,
                     height = 800,
                     max_words = 300,
                     colormap='Set3',
                     min_word_length = 4,
                     margin = 0,
                     max_font_size = 130,
                     random_state = 42,
                     ).generate(str(sorted_tweets['user_description']))
print(wordcloud)
plt.figure(figsize = (20,12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
